Numerical Methods for ODE's

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
"Parameters"
h = 0.013
t = np.arange(0, 1.3, h, dtype = np.double)
x = np.zeros(len(t))
x[0] = -1
```

```python
"IVP1"
f1 = lambda t, x: np.exp(-t)
ivp1_exact = -np.exp(-t)
```

```python
"Explicit Euler Method"
def eem(f):
    for i in range(0, len(t)-1):
        x[i+1] = x[i] + h*f(t[i], x[i])
    return x
exp_eul = eem(f1)
```

```python
"Implicit Euler Method"
def iem(f):
    for i in range(0, len(t) - 1):
        x[i+1] = x[i] + h*f(t[i+1], x[i+1])
    return x
```

```python
"Implicit Midpoint rule Method"
def imrm(f):
    for i in range(0, len(t) - 1):
        x[i+1] = x[i] + h*f(t[i]+(h/2), (x[i+1]+x[i])/2)
    return x
```

```python
"Explicit Modified Euler Method"
def emem(f):
    for i in range(0, len(t) - 1):
        x[i+1] = x[i] + h*f(t[i]+(h/2), x[i] + f(t[i], x[i])*(h/2))
    return x
```

```python
"Trapezoidal Method"
def tm(f):
    for i in range(0, len(t) - 1):
        x[i+1] = x[i] + (h/2)*(f(t[i], x[i]) + f(t[i+1], x[i+1]))
    return x
```

```python
plt.figure(figsize = (18, 12))
plt.title('Explicit Euler Method and Exact Solution for IVP1')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.plot(t, eem(f1),'bo-.', label='Explicit Euler Method')
plt.plot(t, ivp1_exact,'r', label='Exact')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.figure(figsize = (18, 12))
plt.title('Implicit Euler Method and Exact Solution for IVP1')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.plot(t, iem(f1), 'bo-.', label='Implicit Euler Method')
plt.plot(t, ivp1_exact, 'r', label='Exact')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.figure(figsize = (18, 12))
plt.title('Implicit Midpoint rule Method and Exact Solution for IVP1')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.plot(t, imrm(f1), 'bo-.', label='Implicit Midpoint rule Method')
plt.plot(t, ivp1_exact, 'r', label='Exact')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.figure(figsize = (18, 12))
plt.title('Explicit Modified Euler Method and Exact Solution for IVP1')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.plot(t, emem(f1), 'bo-.', label='Explicit Modified Euler Method')
plt.plot(t, ivp1_exact, 'r', label='Exact')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.figure(figsize = (18, 12))
plt.title('Trapezoidal Method and Exact Solution for IVP1')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.plot(t, tm(f1), 'bo-.', label='Trapezoidal Method')
plt.plot(t, ivp1_exact, 'r', label='Exact')
plt.legend(loc='lower right')
plt.show()
```
